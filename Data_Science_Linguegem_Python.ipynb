```python
#for fora da função
idades = [15,18,25,6,28]
def verifica_se_pode_dirigir(idade):
  if(idade >= 18):
    print(f'Já pode dirigir,{idade} maior de idade')
  else:
    print(f'Não pode dirigir, {idade} menor de idade')

for idade in idades:
 verifica_se_pode_dirigir(idade)      
```

```python
#dentro
idades = [25,16,19,25,15,55]
def verifica_se_pode_dirigir(idade):
  for idade in idades:

    if(idade >= 18):
      print(f'Já pode dirigir,{idade} maior de idade')
    else:
      print(f'Não pode dirigir, {idade} menor de idade')


verifica_se_pode_dirigir(idades) 
```

```python
permisao = []
idades = [15,19,17,28]
def permite(idade, permisao):
  for idade in idades:
    if idade >= 18:
      permisao.append(True)
    else:
      permisao.append(False)  


permite(idades,permisao)  
 
```

```python
for p in permisao:
    if p == True:
      print('Pode dirigir ')
    else:
      print('Não dirigir ')
```

```python
notas_matematica = ['Matemática',8,7,6,6,7,7,8,10]
notas_portugues = ['Português',9,9,9,8,5,6,8,5]
notas_geografia = ['Geografia',10,10,6,7,7,7,8,7]

notas = [notas_matematica, notas_portugues, notas_geografia]
```

```python
for nota in notas:
 x = list(range(1, 9))
 y = nota[1:]
 plt.plot(x, y, marker='o')
 plt.xlabel('Provas')
 plt.ylabel('Notas')
 plt.title(nota[0])
 plt.show()
```
