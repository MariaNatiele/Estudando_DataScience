```python
#Importanado as bibliotecas gráficos
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Base dados
dataframe = pd.read_csv("trees.csv")
dataframe.shape
```

```python
dataframe.head(5)
```

```python
#iloc[,1] seleciona a coluna Height - > Altura; bins= 6 -> Quantidade de eixos
histograma = np.histogram(dataframe.iloc[:,1],bins= 6 )
histograma
```

```python
#gerando o gráfico

plt.hist(dataframe.iloc[:,1],bins=6)
plt.title("Histograma")
plt.xlabel("Altura")
plt.ylabel("Frequência")
```

Gráfico de Densidade

```python
#histograma no Seaborn
sns.histplot(dataframe.iloc[:,1],kde=False, bins=6, color="Blue" ).set(title = ("Arvore"))
```

```python
sns.kdeplot(dataframe.iloc[:,1],color="Blue" ).set(title = ("Arvore"))
```

```python
#Kde true -> une o hist kdeplot
sns.histplot(dataframe.iloc[:,1],kde=True, bins=6, color="Blue" ).set(title = ("Arvore"))
```

Gráfico dispersão

```python
#marker="."  aparece como icone no gráfico
plt.scatter(x=dataframe.Girth, y=dataframe.Volume, color="green", facecolors="none", marker="." )
plt.title("Arvore")
plt.xlabel("Circunferencia")
plt.ylabel("Volume")
```

```python
plt.plot(dataframe.Girth,dataframe.Volume)
plt.title("Arvore")
plt.xlabel("Circunferencia")
plt.ylabel("Volume")
```

```python
#gráfico de disperssão no seaborn
#dá aleatoriedade as dados -> x_jitter = 0.3; 
# Linha de tendência -> fit_reg = True
sns.regplot(x=dataframe.Girth, y=dataframe.Volume, data = dataframe, x_jitter = 0.3, fit_reg = True)
```

Gráfico de dispersão com legenda

```python
dados = pd.read_csv("co2.csv")
dados.shape
```

```python
dados.head(5)
```

```python
#selecionando as variaveis
x = dados.conc
y = dados.uptake
```

```python
#lista dos valores unicos de dataframe.Treatment
unicos = list(set(dados.Treatment))
unicos
```

```python

#gráfico
for i in range(len(unicos)):
  indice = dados.Treatment == unicos[i]
  plt.scatter(x[indice], y[indice], label = unicos[1])
plt.legend(loc="lower right")
```

Boxplot

```python
#notch=True -> detalhe proximo a mediana; showfliers -> desconsidera o outilier
plt.boxplot(dataframe.Volume, vert=False, showfliers=False, notch=True, patch_artist=True)
plt.title("Arvore")
plt.xlabel("Volume")
```

#Gráfico 3D

```python
dadosG = pd.read_csv('/content/orchard.csv')
dadosG.shape
```

```python
dadosG.head()
```

```python
from  mpl_toolkits.mplot3d import axes3d
figura  = plt.figure()
eixo = figura.add_subplot(1,1,1, projection = '3d')
eixo.scatter(dadosG.decrease, dadosG.rowpos, dadosG.colpos)
eixo.set_xlabel('decrease')
eixo.set_ylabel('rowpos')
eixo.set_zlabel('colpos')




```
