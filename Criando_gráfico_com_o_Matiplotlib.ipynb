```python
import pandas as pd
```

```python
dados = pd.read_csv('monitoramento_tempo.csv')
```

```python
dados.head()
```

```python
# olhar se a base está completa
dados.info()
```

```python
#transformar string em data
dados['data'] = pd.to_datetime(dados['data'])
```

```python
import matplotlib.pyplot as plt
```

```python
#plota o gráfico
plt.plot(dados['data'],dados['temperatura']);
```

```python
#Aumenta o tamanho do gráfico
plt.figure(figsize=(15,8))
plt.plot(dados['data'],dados['temperatura']);
```

```python
#titulo no gráfico
plt.figure(figsize=(15,8))
plt.plot(dados['data'],dados['temperatura'])
plt.title('Temperatura e tempo');
```

```python
#colocando eixos no grafico
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
```

```python
#colocando a figura no eixo
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'])
eixo.set_title('Temperatura durante os anos');
```

```python
#colocando legenda
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'])
eixo.set_title('Temperatura durante os anos')
eixo.set_ylabel('Temperatura')
eixo.set_xlabel('Data');
```

```python
#aumentando a fonte 
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'])
eixo.set_title('Temperatura durante os anos',fontsize=25)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Data',fontsize=20);
```

```python
#colocando legenda 
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'])
eixo.set_title('Temperatura durante os anos',fontsize=25)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.legend(['temperatura'],loc = 'lower right',fontsize = 15)
```

```python
#alterando a cor suporta cores em Hexadecimal
fig = plt.figure(figsize=(15,8))
eixo = fig.add_axes([0,0,1,1])
eixo.plot(dados['data'],dados['temperatura'],color = '#696969')
eixo.set_title('Temperatura durante os anos',fontsize=25)
eixo.set_ylabel('Temperatura',fontsize=20)
eixo.set_xlabel('Data',fontsize=20)
eixo.legend(['temperatura'],loc = 'lower right',fontsize = 15)
```
